```python
# xが偶数だとx/2，奇数だと3*x+1を計算して返す関数f（本当はもっと単純に書ける）．
def f(x):
    if x % 2 == 0:
        return x / 2
    else: 
        return 3*x + 1
```

```python
f(20)  # 表示されない（表示したい場合は print(f(20)) とする）
f(21)  # 最後の行だけ表示される
```

# ファイルにアクセスしたい

Google Colabは皆さんのパソコンにあるファイルにアクセスできません（別の計算機なの当然）．
Google Colabがアクセスできるのは，**Google Drive**にあるファイルです．

Google Driveは「インターネット上の自分専用のフォルダ」くらいの認識でいいです．そこにファイルを保存します．

```python
# Google Driveをマウント（読み込み）
from google.colab import drive
drive.mount('/content/drive/')

# 上を実行すると，
## (1) リンクが出てくるのでそれをクリック，
##（2）Googleアカウントにログインして認証コードが出てくる，
##（3）コードをコピーして表示されているボックスに貼り付け．

# 目的の場所（フォルダ・ディレクトリ）へ移動（各自の環境で適宜修正）
%cd "/content/drive/MyDrive/Colab Notebooks/JKJ1A"
# contentフォルダの中のdriveフォルダの中のMyDriveの中の....JKJ1Aに移動している
```
